```python
from typing import Union, Sequence
import numpy as np

def print_arr(
    *arr: Sequence[Union[torch.Tensor, np.ndarray]],
    prefix: str = ""
) -> None:  # âœ… Correct placement of return type hint
    print(
        "\n\n".join(
            f"{prefix}{str(x)} <shape: {x.shape}> <dtype: {x.dtype}>"
            for x in arr
        )
    )

t = torch.rand((3, 3), dtype=torch.float32)
print_arr(t, prefix='My Tensor = ')
```
