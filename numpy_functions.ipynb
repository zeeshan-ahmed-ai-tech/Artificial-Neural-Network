```python
# ===============================
# NumPy Array Creation and Operations
# ===============================

import numpy as np

# --- Creating Arrays ---
a = np.array([1, 2, 3])
b = np.array([(1.5, 2, 3), (4, 5, 6)], dtype=float)
c = np.array([[(1.5, 2, 3), (4, 5, 6)], [(3, 2, 1), (4, 5, 6)]], dtype=float)

print("Array a:\n", a)
print("\nArray b:\n", b)
print("\nArray c:\n", c)

# --- Array Initialization ---
print("\nZeros array:\n", np.zeros((3, 4)))                   # Create an array of zeros
print("\nOnes array:\n", np.ones((2, 3, 4), dtype=np.int16))   # Create an array of ones
d = np.arange(10, 25, 5)                                       # Evenly spaced values (step = 5)
print("\nArray d (arange):\n", d)
print("\nLinspace array:\n", np.linspace(0, 2, 9))             # Evenly spaced samples
e = np.full((2, 2), 7)                                         # Constant array
print("\nFull array (7s):\n", e)
f = np.eye(2)                                                  # 2x2 Identity matrix
print("\nIdentity matrix f:\n", f)
print("\nRandom array:\n", np.random.random((2, 2)))           # Random values
print("\nEmpty array:\n", np.empty((3, 2)))                    # Empty array (uninitialized values)

# --- Arithmetic Operations ---
print("\nSubtraction (a - b):\n", a - b)
print("\nSubtraction using np.subtract:\n", np.subtract(a, b))
print("\nAddition (b + a):\n", b + a)
print("\nAddition using np.add:\n", np.add(b, a))
print("\nDivision (a / b):\n", a / b)
print("\nDivision using np.divide:\n", np.divide(a, b))
print("\nMultiplication (a * b):\n", a * b)
print("\nMultiplication using np.multiply:\n", np.multiply(a, b))

# --- Mathematical Functions ---
print("\nExponential of b:\n", np.exp(b))
print("\nSquare root of b:\n", np.sqrt(b))
print("\nSine of a:\n", np.sin(a))
print("\nCosine of b:\n", np.cos(b))
print("\nNatural log of a:\n", np.log(a))

# --- Dot Product ---
print("\nDot product of e and f:\n", e.dot(f))
```
