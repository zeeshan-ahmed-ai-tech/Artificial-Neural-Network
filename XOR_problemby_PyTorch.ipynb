```python
import torch
import torch.nn as nn

# ðŸŸ¢ Step 1: Dummy data (2 input values and 1 output value)
X = torch.tensor([[1.0], [2.0], [3.0], [4.0]])   # input
y = torch.tensor([[2.0], [4.0], [6.0], [8.0]])   # output (2x relation)

# ðŸŸ¢ Step 2: Simple model (1 input neuron â†’ 1 hidden layer â†’ 1 output neuron)
class SimpleANN(nn.Module):
    def __init__(self):
        super(SimpleANN, self).__init__()
        self.layer1 = nn.Linear(1, 1)   # linear layer: y = wx + b

    def forward(self, x):
        return self.layer1(x)           # forward pass

# ðŸŸ¢ Step 3: Create model, define loss & optimizer
model = SimpleANN()
criterion = nn.MSELoss()                # Mean Squared Error (for regression)
optimizer = torch.optim.SGD(model.parameters(), lr=0.01)

# ðŸŸ¢ Step 4: Training loop
for epoch in range(1000):
    y_pred = model(X)                   # forward pass
    loss = criterion(y_pred, y)         # calculate loss
    loss.backward()                     # backpropagation
    optimizer.step()                    # update weights
    optimizer.zero_grad()               # reset gradients

    if (epoch+1) % 100 == 0:
        print(f'Epoch {epoch+1}, Loss: {loss.item():.4f}')

# ðŸŸ¢ Step 5: Test / Predict
with torch.no_grad():
    test = torch.tensor([[5.0]])
    prediction = model(test)
    print("\nPredicted value for input 5.0:", prediction.item())
```
